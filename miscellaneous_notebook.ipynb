```python
movie_db_info = pd.read_sql("""

SELECT  DISTINCT movie_id, primary_title, original_title, start_year, runtime_minutes, averagerating, numvotes, primary_name, primary_profession, genres
FROM movie_basics
JOIN movie_ratings
    USING(movie_id)
JOIN directors_no_dups
    USING(movie_id)
JOIN writers_no_dups
    USING(movie_id)

""", conn)

movie_db_info.info()
```

```python
movie_db_info.isna().sum()
```

```python
movie_db_info.dropna(subset=['runtime_minutes', 'primary_profession', 'genres'], inplace=True)
```
